```python
import deepchem as dc
```

```python
dataset_file = 'molnet_bace.csv'
tasks = ['pIC50']
featurizer_func = dc.feat.ConvMolFeaturizer()

loader = dc.data.CSVLoader(tasks=tasks, feature_field='mol', featurizer=featurizer_func)
dataset = loader.create_dataset(dataset_file)

transformer = dc.trans.BalancingTransformer(dataset=dataset)
dataset = transformer.transform(dataset)

splitter = dc.splits.RandomSplitter()
train_dataset, test_dataset = splitter.train_test_split(dataset)
```

```python
n_tasks = len(tasks)
model = dc.models.GraphConvModel(n_tasks, mode='regression')
model.fit(train_dataset, nb_epoch=150)
```

```python
pred_train_y = model.predict(train_dataset)
pred_test_y = model.predict(test_dataset)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import r2_score
```

```python
gnn_train_mae = round(mean_absolute_error(train_dataset.y, pred_train_y), 2)
gnn_train_mse = round(mean_squared_error(train_dataset.y, pred_train_y), 2)
gnn_train_mape = round(mean_absolute_percentage_error(train_dataset.y, pred_train_y), 2)
gnn_train_r2 = round(r2_score(train_dataset.y, pred_train_y), 2)

gnn_test_mae = round(mean_absolute_error(test_dataset.y, pred_test_y), 2)
gnn_test_mse = round(mean_squared_error(test_dataset.y, pred_test_y), 2)
gnn_test_mape = round(mean_absolute_percentage_error(test_dataset.y, pred_test_y), 2)
gnn_test_r2 = round(r2_score(test_dataset.y, pred_test_y), 2)
```

```python
gnn_perf = {'mae':[gnn_train_mae, gnn_test_mae],
            'mse':[gnn_train_mse, gnn_test_mse],
            'mape':[gnn_train_mape, gnn_test_mape],
            'r2':[gnn_train_r2, gnn_test_r2]}
```

```python
gnn_perf
```
